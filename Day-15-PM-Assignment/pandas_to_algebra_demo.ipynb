# Day 15 PM — Pandas to Relational Algebra (Part A)

Map 3 Pandas operations to relational algebra.

```python
import pandas as pd

# Load sample OrderFacts
df = pd.read_csv('OrderFacts_sample.csv')
orders = df[['order_id', 'customer_id', 'restaurant_id', 'order_date']].drop_duplicates()
customers = df[['customer_id', 'customer_name', 'customer_phone']].drop_duplicates()
order_items = df[['order_id', 'item_id', 'quantity', 'unit_price']]

# 1. merge ~ Join (⋈)
merged = orders.merge(customers, on='customer_id')
print('Merge (⋈): orders ⋈ customers')
print(merged.head())

# 2. groupby + agg ~ Group and aggregate
totals = order_items.groupby('order_id').agg(total_qty=('quantity', 'sum'))
print('\nGroupBy+agg: G_order_id sum(quantity)(OrderItem)')
print(totals)

# 3. filter ~ Selection (σ)
filtered = orders[orders['order_id'] == 1]
print('\nFilter: σ_order_id=1(Order)')
print(filtered)
```
