# Day 16 PM — EDA Report: Data Visualization

Comprehensive EDA on **eda_assignment_data.csv** (sample; real file on LMS has 1200 rows, 10 mixed-type columns).

Plots: 3 distribution (hist+KDE), 2 relationship (scatter+regression), 1 correlation heatmap, 1 box plot. Each with title, labelled axes, and 2-sentence insight.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('eda_assignment_data.csv')
df.head()
```

## 1. Distribution plots (histogram with KDE) — 3 numerical features

**Insight:** Sales and quantity are right-skewed; price is more spread. The KDE shows the continuous shape of the distribution, which helps assess normality and outliers.

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for ax, col in zip(axes, ['sales', 'quantity', 'price']):
    sns.histplot(df[col], kde=True, ax=ax, bins=30)
    ax.set_title(f'Distribution of {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

## 2. Relationship plots — scatter with regression line (2 pairs)

**Insight:** Regplot shows the linear trend and confidence interval. Price vs sales and quantity vs sales help identify whether higher price or volume correlates with revenue.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.regplot(data=df, x='price', y='sales', ax=axes[0], scatter_kws={'alpha':0.4})
axes[0].set_title('Sales vs Price (with regression line)')
axes[0].set_xlabel('Price')
axes[0].set_ylabel('Sales')
sns.regplot(data=df, x='quantity', y='sales', ax=axes[1], scatter_kws={'alpha':0.4})
axes[1].set_title('Sales vs Quantity (with regression line)')
axes[1].set_xlabel('Quantity')
axes[1].set_ylabel('Sales')
plt.tight_layout()
plt.show()
```

## 3. Correlation heatmap — all numerical features, annotated

**Insight:** The heatmap shows which numeric variables correlate (e.g. quantity–sales). Annotated values make it easy to report exact correlation coefficients for reporting.

```python
num = df.select_dtypes(include=['number'])
corr = num.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0, square=True)
plt.title('Correlation heatmap (numerical features)')
plt.tight_layout()
plt.show()
```

## 4. Box plot — key metric across categories

**Insight:** Sales distribution differs by category; some categories have higher median and spread. The box plot highlights outliers and quartiles for each group.

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='category', y='sales')
plt.title('Sales by category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig('dashboard_boxplot.png', dpi=120, bbox_inches='tight')
plt.show()
```
