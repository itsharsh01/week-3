```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Part A — 5 scenarios
print('1. Website (Poisson):', round(1 - stats.poisson.cdf(220, 200), 4))
print('2. Quality (Binomial):', round(stats.binom.pmf(3, 50, 0.02), 4))
print('3. Delivery (Normal):', round(1 - stats.norm.cdf(60, 45, 8), 4), ',', round(stats.norm.cdf(50, 45, 8) - stats.norm.cdf(40, 45, 8), 4))
print('4. Arrivals (Poisson):', round(stats.poisson.pmf(0, 1), 4))
print('5. CLT: X̄ ~ N(μ, σ²/35)')
```

```python
# Part B: Beta PDFs
x = np.linspace(0.001, 0.999, 200)
for a, b, label in [(2, 5, 'Beta(2,5)'), (5, 5, 'Beta(5,5)'), (0.5, 0.5, 'Beta(0.5,0.5)')]:
    plt.plot(x, stats.beta.pdf(x, a, b), label=label)
plt.legend(); plt.xlabel('p'); plt.title('Beta PDFs'); plt.show()

# Posterior: 7 heads, 3 tails from Beta(1,1) -> Beta(8,4)
plt.plot(x, stats.beta.pdf(x, 8, 4), label='Posterior Beta(8,4)')
plt.legend(); plt.xlabel('p'); plt.title('After 7H, 3T'); plt.show()
```

```python
# Part C Q2: simulate_clt with overlay
def simulate_clt(distribution, params, n_samples, n_simulations):
    dist = distribution(**params)
    means = [np.mean(dist.rvs(size=n_samples)) for _ in range(n_simulations)]
    theo_mean = dist.mean()
    theo_std = np.sqrt(dist.var() / n_samples)
    return np.array(means), theo_mean, theo_std

means, m, s = simulate_clt(stats.expon, {'scale': 2}, 30, 5000)
plt.hist(means, bins=40, density=True, alpha=0.7, label='Sample means')
xx = np.linspace(means.min(), means.max(), 100)
plt.plot(xx, stats.norm.pdf(xx, m, s), 'r-', lw=2, label='Theoretical N')
plt.legend(); plt.title('CLT: Exponential(2), n=30'); plt.show()
```
